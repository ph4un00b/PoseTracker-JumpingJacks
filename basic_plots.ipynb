```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Generate mocking data
np.random.seed(0)
num_hospitals = 100
total_discharges = np.random.randint(50, 500, num_hospitals)
average_medicare_payments = np.random.uniform(1000, 10000, num_hospitals)

# Create DataFrame
data = {
    'Total Discharges': total_discharges,
    'Average Medicare Payments': average_medicare_payments
}
df = pd.DataFrame(data)
df.head()
```

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Total Discharges'], df['Average Medicare Payments'], color='blue', alpha=0.5)
plt.title('Relationship between Total Discharges and Average Medicare Payments')
plt.xlabel('Total Discharges')
plt.ylabel('Average Medicare Payments')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 🔥 line plot

```python
# prompt: create a line plot, mock the dataset to be records of  passengers onboards in a airport over time

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Generate mocking data
np.random.seed(0)
num_months = 12
passengers = np.random.randint(100, 500, num_months)
months = range(1, num_months + 1)

# Create DataFrame
data = {
    'Month': months,
    'Passengers': passengers
}
df = pd.DataFrame(data)

print(df.head())
# Create line plot
# plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Passengers'], color='blue', linewidth=2)
plt.title('Number of Passengers over Time')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 🔥 Distribution plots (look at dispersed data)

```python
# prompt: use de total bill and tip correlation dataset to create an histogram, mock the dataset

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Generate mocking data
np.random.seed(0)
num_restaurants = 100
total_bills = np.random.uniform(10, 100, num_restaurants)
tip_increases = np.random.uniform(0, 50, num_restaurants)

# Create DataFrame
data = {
    'Total Bill': total_bills,
    'Tip': tip_increases
}
df = pd.DataFrame(data)

print(df.head())
# Create histogram for total bill
plt.hist(df['Total Bill'], bins=10, color='blue', alpha=0.5)
plt.title('Distribution of Total Bills')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()

# Create histogram for tip increase
plt.hist(df['Tip'], bins=20, color='blue', alpha=0.5)
plt.title('Distribution of Tip Increases')
plt.xlabel('Tip Increase')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# prompt: use de total bill and tip correlation dataset to create an box-and-whisker plot, mock the dataset

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Generate mocking data
np.random.seed(0)
num_restaurants = 100
total_bills = np.random.uniform(10, 100, num_restaurants)
tip_increases = np.random.uniform(0, 50, num_restaurants)

# Create DataFrame
data = {
    'Total Bill': total_bills,
    'Tip': tip_increases
}
df = pd.DataFrame(data)

print(df.head())
# Create box-and-whisker plot
plt.figure(figsize=(10, 6))
plt.boxplot([df['Total Bill'], df['Tip']], labels=['Total Bill', 'Tip'], patch_artist=True)
plt.title('Box-and-Whisker Plot of Total Bill and Tip')
plt.xlabel('Variable')
plt.ylabel('Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```
